```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generar datos
X_blobs, y_blobs = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Visualizar datos
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs, cmap='rainbow', edgecolor='k')
plt.title("Datos generados con make_blobs")
plt.show()
```

```python
from sklearn.datasets import make_moons

# Generar datos
X_moons, y_moons = make_moons(n_samples=300, noise=0.1, random_state=42)

# Visualizar datos
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap='rainbow', edgecolor='k')
plt.title("Datos generados con make_moons")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Elegir un conjunto de datos (por ejemplo, make_moons)
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar SVM con kernel lineal
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
print("Precisión SVM (kernel lineal):", accuracy_score(y_test, y_pred_linear))

# Entrenar SVM con kernel RBF
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
print("Precisión SVM (kernel RBF):", accuracy_score(y_test, y_pred_rbf))
```
